```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns

nobel_df = pd.read_csv("nobel_laureates.csv")
top5 = nobel_df.groupby('Category').size().sort_values(ascending=False).head(5)

sns.set_style('dark')
icons = ['medicina.png', 'fisica.png', 'quimica.png', 'peace.png', 'literatura.png']
sizes = top5.values
total_size = sum(sizes)
bottom = [-50, -50, -50, 0, 20]


fig, ax = plt.subplots(figsize=(8, 8))
for i, icon in enumerate(icons):
    img = mpimg.imread(icon)
    img_size = sizes[i] * 1.5
    ax.imshow(img, extent=[i - 0.5, i + 0.5, bottom[i] + 100, img_size], aspect='auto')
    ax.text(i, img_size + 10, sizes[i], ha='center', va='center', fontsize=16)
    
ax.set_xticks(range(5))
ax.set_yticks([])
ax.set_xticklabels(top5.index.tolist(), fontsize=16, rotation=45, ha='right')
ax.set_xlim([-0.5, 4.5])
ax.set_ylim([0, 500])
ax.set_title('Top 5 Disciplines with the Most Nobel Prizes', fontsize=20)
plt.show()
```
